```python
import selenium
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests
import pprint
import re
```

```python
url = []
cover_image = []
what_will_learn = []
post_course_interaction = []
international_faculty = []
human_interaction = []
personalised_teaching = []
live_class = []
job_assistance = []
internship = []
alumni_network = []
learning_medium = []
virtual_labs = []
case_based_learning = []
capstone = []
Faq1ques = []
Faq1ans = []
```

```python
driver = selenium.webdriver.Chrome(r'C:\Users\91975\Downloads\chromedriver_win32 (1)/chromedriver.exe')
```

```python
for i in df['Url']:
    
    driver.get(i)
    res1 = driver.page_source.encode('utf-8')
    soup1 = BeautifulSoup(res1, "html.parser")
    
    
    #url
    url.append(i)
    
    #cover image
    k = soup1.find_all('div',{'class':'course-hero_relatedNd__1TpBm'})
    if len(k)>0:
        for l in k:
            if (l!= None):
                cover_image.append(l.get('style'))

    if len(k)==0:
        cover_image.append(' ')
            
            
    #what will learn
    w = soup1.find_all('div',{'class':'course-syllabus_syllabusList__ZbR2Q'})
    if len(w)>0:
        for l in w:
            if (l.text!= None):
                what_will_learn.append(l.text)

    if len(w)==0:
        what_will_learn.append(' ')
            
            
    #post course interaction
    the_word1 = 'Post Course Interaction'
    try:
        words = soup1.find_all(text=lambda text: text and the_word1 in text)
        if the_word1 in words:
            post_course_interaction.append(True)
        else:
            post_course_interaction.append(False)
    except:
        post_course_interaction.append(' ')
            
        
    #international faculty
    international_faculty.append(True)

    #human interaction
    the_word2 = 'Human Interaction'
    try:
        words = soup1.find_all(text=lambda text: text and the_word2 in text)
        if the_word2 in words:
            human_interaction.append(True)
        else:
            human_interaction.append(False)
    except:
        human_interaction.append(' ')

    #personalised teaching
    the_word3 = 'Customised Teaching'
    try:
        words = soup1.find_all(text=lambda text: text and the_word3 in text)
        if the_word3 in words:
            personalised_teaching.append(True)
        else:
            personalised_teaching.append(False)
    except:
        personalised_teaching.append(' ')

    #live class
    the_word4 = 'Live Class'
    try:
        words = soup1.find_all(text=lambda text: text and the_word4 in text)
        if the_word4 in words:
            live_class.append(True)
        else:
            live_class.append(False)
    except:
        live_class.append(' ')

    #job assistance
    the_word5 = 'Job Assistance'
    try:
        words = soup1.find_all(text=lambda text: text and the_word5 in text)
        if the_word5 in words:
            job_assistance.append(True)
        else:
            job_assistance.append(False)
    except:
        job_assistance.append(' ')

    #internship
    the_word6 = 'Internship'
    try:
        words = soup1.find_all(text=lambda text: text and the_word6 in text)
        if the_word6 in words:
            internship.append(True)
        else:
            internship.append(False)
    except:
        internship.append(' ')

    #alumni network
    the_word7 = 'Alumni'
    try:
        words = soup1.find_all(text=lambda text: text and the_word7 in text)
        if the_word7 in words:
            alumni_network.append(True)
        else:
            alumni_network.append(False)
    except:
        alumni_network.append(' ')

    #learning medium
    learning_medium.append('Instructor videos')

    #virtual lab
    the_word8 = 'Virtual lab'
    try:
        words = soup1.find_all(text=lambda text: text and the_word8 in text)
        if the_word8 in words:
            virtual_labs.append(True)
        else:
            virtual_labs.append(False)
    except:
        virtual_labs.append(' ')

    #case based learning
    the_word9 = 'Case study'
    try:
        words = soup1.find_all(text=lambda text: text and the_word9 in text)
        if the_word9 in words:
            case_based_learning.append(True)
        else:
            case_based_learning.append(False)
    except:
        case_based_learning.append(' ')


    #capstone
    the_word10 = 'Capstone'
    try:
        words = soup1.find_all(text=lambda text: text and the_word10 in text)
        if the_word10 in words:
            capstone.append(True)
        else:
            capstone.append(False)
    except:
        capstone.append(' ')

    #faq ques
    Faq1ques.append('Why Take This Course')

    #faq ans
    w = soup1.find_all('div',{'class':'why-taken_reason__3x0Wq'})
    if len(w)>0:
        for l in w:
            if (l.text!= None):
                Faq1ans.append(l.text)

    if len(w)==0:
        Faq1ans.append(' ')
```

```python
data={'Url':url,
    'cover_image':cover_image,
'what_will_learn':what_will_learn,
'post_course_interaction':post_course_interaction,
'human_interaction':human_interaction,
'personalised_teaching':personalised_teaching,
'live_class':live_class,
'job_assistance':job_assistance,
'internship':internship,
'alumni_network':alumni_network,
'learning_medium':learning_medium,
'virtual_labs':virtual_labs,
'case_based_learning':case_based_learning,
'capstone_project':capstone,
'Faq1ques':Faq1ques,
'Faq1ans':Faq1ans
}

df2 = pd.DataFrame.from_dict(data, orient='index')
df2 = df2.transpose()
```
