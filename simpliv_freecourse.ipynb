```python
import selenium
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests
import pprint
import re
```

```python
df = pd.read_csv('simpliv.xlsx - Sheet1.csv')
```

```python
url = []
cover_video = []
post_course_interaction = []
alumni_network = []
capstone = []
```

```python
driver = selenium.webdriver.Chrome(r'C:\Users\91975\Downloads\chromedriver_win32 (1)/chromedriver.exe')
```

```python
for i in df['partner_course_url']:
    
    driver.get(i)
    res1 = driver.page_source.encode('utf-8')
    soup1 = BeautifulSoup(res1, "html.parser")
    
    
    #url
    url.append(i)
    
    #cover video
    k = soup1.find_all('div',{'class':'preview-video-wrapper'})
    if len(k)>0:
        for l in k:
            if (l!= None):
                cover_video.append(l.get('src'))

    if len(k)==0:
        cover_video.append(' ')
    
    
    #post course interaction
    the_word1 = 'Post Course Interaction'
    try:
        words = soup1.find_all(text=lambda text: text and the_word1 in text)
        if the_word1 in words:
            post_course_interaction.append(True)
        else:
            post_course_interaction.append(False)
    except:
        post_course_interaction.append(' ')
        
    
    the_word7 = 'Alumni'
    try:
        words = soup1.find_all(text=lambda text: text and the_word7 in text)
        if the_word7 in words:
            alumni_network.append(True)
        else:
            alumni_network.append(False)
    except:
        alumni_network.append(' ')
        
        
    #capstone
    the_word10 = 'Capstone'
    try:
        words = soup1.find_all(text=lambda text: text and the_word10 in text)
        if the_word10 in words:
            capstone.append(True)
        else:
            capstone.append(False)
    except:
        capstone.append(' ')
```

```python
data={'url':url,
'post_course_interaction':post_course_interaction,
'alumni_network':alumni_network,
'capstone_project':capstone,
}
```

```python
df2 = pd.DataFrame.from_dict(data, orient='index')
df2 = df2.transpose()
```
